```python
import pandas as pd
import numpy as np
data=pd.read_csv("Train_data.csv")
data
```

```python
import numpy as np
from sklearn import svm
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X = data.loc[:, 'dst_host_diff_srv_rate':'dst_host_srv_rerror_rate']
y = data.loc[:, 'class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
clf = svm.SVC()
clf.fit(X_train_imputed, y_train)
y_pred = clf.predict(X_test_imputed)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
support_vectors = clf.support_vectors_
print(f'Support Vectors: {support_vectors}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

data = pd.read_csv("Train_data.csv")
class1 = "normal"
class2 = "anomaly"
selected_data = data[(data['class'] == class1) | (data['class'] == class2)]
selected_data = pd.get_dummies(selected_data, columns=['protocol_type', 'service', 'flag'])

X = selected_data.drop(columns=['class'])
y = selected_data['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X_train, y_train)

accuracy = svm_model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
X = data.loc[:, 'dst_host_diff_srv_rate':'dst_host_srv_rerror_rate']
y = data.loc[:, 'class']

categ_var = data.select_dtypes(include=['object'])
label_encoder = LabelEncoder()
for col in categ_var:
    data[col] = label_encoder.fit_transform(data[col])
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3, random_state=42)
svm = SVC(kernel='linear', C=1)
svm.fit(X_train, y_train)
predicted_labels = svm.predict(X_test)
decision_values = svm.decision_function(X_test)
accuracy = (predicted_labels == y_test).mean()
print("Predicted Labels:", predicted_labels)
print("\nTrue Labels:", y_test)
print("\nDecision Function Values:", decision_values)
print(f"\nAccuracy: {accuracy * 100:.2f}%")
```
